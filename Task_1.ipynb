```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('StudentPerformance.csv')
df.head()
```

```python
df.info()
df.describe()
```

```python
avg_performance = df['Performance Index'].mean()
print("Average Performance Index:", avg_performance)
```

```python
df.groupby('Extracurricular Activities')['Performance Index'].mean().plot(kind='bar')
plt.title('Average Performance vs Extracurricular Activities')
plt.xlabel('Extracurricular Activities')
plt.ylabel('Average Performance Index')
plt.show()
```

```python
plt.scatter(df['Hours Studied'], df['Performance Index'])
plt.xlabel('Hours Studied')
plt.ylabel('Performance Index')
plt.title('Hours Studied vs Performance Index')
plt.show()
```

```python
plt.figure(figsize=(8,6))

numeric_df = df.select_dtypes(include='number')
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')

plt.title('Heatmap')
plt.show()
```

```python
print("Observations from Data Analysis:")
print("1. Performance Index increases as Hours Studied increases.")
print("2. Previous Scores have a strong positive impact on Performance Index.")
print("3. Students who practice more sample question papers perform better.")
print("4. Adequate sleep hours contribute to improved performance.")
print("5. Extracurricular activities show variation in average performance.")
```
