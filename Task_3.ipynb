```python
import numpy as np
print("NumPy imported successfully.")
```

```python
def get_matrix_input():
    """
    Prompts the user to enter the dimensions and elements of a matrix,
    returning it as a NumPy array.
    """
    while True:
        try:
            num_rows = int(input("Enter the number of rows: "))
            num_cols = int(input("Enter the number of columns: "))
            if num_rows <= 0 or num_cols <= 0:
                print("Number of rows and columns must be positive integers. Please try again.")
                continue
            break
        except ValueError:
            print("Invalid input. Please enter integers for dimensions.")

    matrix_rows = []
    print(f"Enter {num_cols} elements for each of the {num_rows} rows, separated by spaces:")
    for i in range(num_rows):
        while True:
            try:
                row_str = input(f"Enter elements for row {i + 1}: ")
                row_elements = [float(x) for x in row_str.split()]
                if len(row_elements) != num_cols:
                    print(f"Invalid number of elements. Expected {num_cols}, but got {len(row_elements)}. Please try again.")
                    continue
                matrix_rows.append(row_elements)
                break
            except ValueError:
                print("Invalid input. Please enter numeric values separated by spaces.")

    return np.array(matrix_rows)

print("Function 'get_matrix_input()' defined successfully.")
```

```python
def add_matrices(matrix_a, matrix_b):
    """
    Performs addition of two matrices, ensuring they have compatible dimensions.
    """
    if matrix_a.shape != matrix_b.shape:
        print("Error: Matrices must have the same dimensions for addition.")
        return None

    return matrix_a + matrix_b

print("Function 'add_matrices()' defined successfully.")
```

```python
def subtract_matrices(matrix_a, matrix_b):
    """
    Performs subtraction of two matrices, ensuring they have compatible dimensions.
    """
    if matrix_a.shape != matrix_b.shape:
        print("Error: Matrices must have the same dimensions for subtraction.")
        return None

    return matrix_a - matrix_b

print("Function 'subtract_matrices()' defined successfully.")
```

```python
def multiply_matrices(matrix_a, matrix_b):
    """
    Performs multiplication of two matrices, ensuring they have compatible dimensions.
    """
    if matrix_a.shape[1] != matrix_b.shape[0]:
        print("Error: The number of columns in the first matrix must equal the number of rows in the second matrix for multiplication.")
        return None

    return np.dot(matrix_a, matrix_b)

print("Function 'multiply_matrices()' defined successfully.")
```

```python
def transpose_matrix(matrix):
    """
    Computes the transpose of a given matrix.
    """
    return np.transpose(matrix)

print("Function 'transpose_matrix()' defined successfully.")
```

```python
def determinant_matrix(matrix):
    """
    Computes the determinant of a given square matrix.
    """
    if matrix.shape[0] != matrix.shape[1]:
        print("Error: Matrix must be square to compute the determinant.")
        return None

    return np.linalg.det(matrix)

print("Function 'determinant_matrix()' defined successfully.")
```

```python
def main_menu():
    """
    Presents a main interactive command-line interface for matrix operations.
    """
    print("\nHello Guys Welcome to the Interactive Matrix Operations Tool!")
    while True:
        print("\nSelect an operation:")
        print("1. Matrix Addition")
        print("2. Matrix Subtraction")
        print("3. Matrix Multiplication")
        print("4. Matrix Transpose")
        print("5. Matrix Determinant")
        print("6. Exit")

        choice = input("Enter your choice (1-6): ")

        if choice == '1':
            print("\n--- Matrix Addition ---")
            print("Enter details for the first matrix:")
            matrix1 = get_matrix_input()
            if matrix1 is None: continue
            print("Enter details for the second matrix:")
            matrix2 = get_matrix_input()
            if matrix2 is None: continue

            result = add_matrices(matrix1, matrix2)
            if result is not None:
                print("\nResult of Addition:\n", result)

        elif choice == '2':
            print("\n--- Matrix Subtraction ---")
            print("Enter details for the first matrix:")
            matrix1 = get_matrix_input()
            if matrix1 is None: continue
            print("Enter details for the second matrix:")
            matrix2 = get_matrix_input()
            if matrix2 is None: continue

            result = subtract_matrices(matrix1, matrix2)
            if result is not None:
                print("\nResult of Subtraction:\n", result)

        elif choice == '3':
            print("\n--- Matrix Multiplication ---")
            print("Enter details for the first matrix:")
            matrix1 = get_matrix_input()
            if matrix1 is None: continue
            print("Enter details for the second matrix:")
            matrix2 = get_matrix_input()
            if matrix2 is None: continue

            result = multiply_matrices(matrix1, matrix2)
            if result is not None:
                print("\nResult of Multiplication:\n", result)

        elif choice == '4':
            print("\n--- Matrix Transpose ---")
            print("Enter details for the matrix to transpose:")
            matrix = get_matrix_input()
            if matrix is None: continue

            result = transpose_matrix(matrix)
            print("\nResult of Transpose:\n", result)

        elif choice == '5':
            print("\n--- Matrix Determinant ---")
            print("Enter details for the matrix to compute determinant:")
            matrix = get_matrix_input()
            if matrix is None: continue

            result = determinant_matrix(matrix)
            if result is not None:
                print("\nResult of Determinant: ", result)

        elif choice == '6':
            print("Exiting the Matrix Operations Tool. Goodbye!")
            break

        else:
            print("Invalid choice. Please enter a number between 1 and 6.")

print("Interactive matrix tool ready.")
```
